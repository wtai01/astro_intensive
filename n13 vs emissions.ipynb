```python
%matplotlib inline
from astropy.io import fits
from astropy import units as u
import os
import pickle as pickle
from astropy.stats import sigma_clipped_stats
from astropy.time import Time
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv (r"C:\Users\taiwe\Documents\Sources Info Sheet - Molecular Catalog.csv")


##CO2 Flux:
df = pd.DataFrame(data, columns= ['Star','n13_30_flux','CO2_flux'])
df_loc = df.loc[df['n13_30_flux'] != -99]
df_loc = df.loc[df['CO2_flux'] != -99]
print (df_loc)
```

```python
star_list = df_loc['Star'].tolist()
n13_30_flux = df_loc['n13_30_flux'].tolist()
CO2_flux = df_loc['CO2_flux'].tolist()
```

```python
fig=plt.figure(1,figsize=(18,10))
ax1=fig.add_subplot(111)
ax1.plot(n13_30_flux, CO2_flux, 'ro')
for i, (x, y) in enumerate(zip(n13_30_flux, CO2_flux)):
    ax1.text(x,y,star_list[i],fontsize=10)
ax1.set_xlabel('n13-30 Flux',fontsize=18)
ax1.set_ylabel('CO2 Flux',fontsize=18)

#ax1.yscale('log')
```
